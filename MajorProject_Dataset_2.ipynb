# 4. Questioning on Refined Dataset

```python
# 1. What is the ID of studio with Highest rank

max_row = df['rank_in_year'].idxmin()  # Finding row in 'rank_in_year' column

highest = df.loc[max_row, 'id']    # Extracting the 'id' from selected row
print(highest)
```

```python
# 2. What is the price of studio with Bedrooms = 3.000000	and Bathrooms = 2.250000 (minimum - 5)

filtered_df = df[(df['bedrooms']==3.000000) & (df['bathrooms'] == 2.250000)]      # Applying Condition
filtered_df[['studio', 'price']].head()        # Printing the Studio and Price
```

```python
# 3. State the studio with Highest rating


max_row = df['rating'].idxmax()  # Finding row in 'rating' column

highest = df.loc[max_row, 'studio']    # Extracting the 'studio' from selected row
print(highest)
```

```python
# 4. State the Worldwide_gross in recent year


max_row = df['year'].idxmax()     # Finding row in 'year' column

highest = df.loc[max_row, 'worldwide_gross']   # Extracting the 'worldwide_gross' from selected row
print(highest)
```

```python
df.query('worldwide_gross == 700059566')
```

```python
# 5. State the year of built for most popular Main Genre


most_popular_genre = df.loc[df['rank_in_year'].idxmin(), 'main_genre']    # Finding the most popular Main Genre (assuming popularity is based on the lowest rank in the year)


filtered_df = df[df['main_genre'] == most_popular_genre]      # Filtering the DataFrame to find the row with the most popular Main Genre


year_of_built = filtered_df['yr_built'].iloc[0]     # Extracting the 'year_of_built' for the most popular Main Genre. Assuming one entry per genre
print(year_of_built)
```
